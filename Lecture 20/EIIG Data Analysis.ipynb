```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

<font size = "5">

Import Data

```python
# Import the raw data from the shared spreadsheets

raw_data = pd.read_csv(r"C:\Users\solrf\Documents\GitHub\emoryqtm151\Lecture 20\data_raw\raw_data.csv")
```

<font size = "5">

Data Cleaning

```python
raw_data.dtypes
```

```python
### Calculate the mean case and overall applicant score for each applicant
data_agg = (raw_data.groupby("Applicant Name")
                      .agg(mean_case_score = ('Case Interview Score','mean'),
                           mean_app_score =   ('Score from 1 to 5 (1 is lowest)','mean')))
```

<font size = "5">

Regression for Case Scores vs Applicant Scores

```python
x=data_agg['mean_case_score'].to_numpy()
x=x.reshape(-1, 1)
y=data_agg['mean_app_score'].to_numpy()
y=y.reshape(-1, 1)

model=LinearRegression().fit(x,y)
y_pred=model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')

plt.xlabel("Average Case Score")
plt.ylabel("Average Overall Applicant Score")

plt.show()
```

<font size = "5">

Statistics for Linear Regression Model


```python
r_squared = model.score(x, y)
print(f"R-Squared: {r_squared}")

print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")



```
